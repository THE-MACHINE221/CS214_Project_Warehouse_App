```python
import numpy as np 
import pandas as pd , re, unidecode
```

```python
product = pd.read_csv("ProductsData.csv", encoding="ISO-8859-1")
```

```python
product.head()
```

```python
product.drop(['product_category', 'professional_publication' , 'region_address' , 'local_address'] , axis=1 , inplace=True)

product.head(10)
```

```python
product.loc[:,'price'].isna().sum()
```

```python
product['price'] = (
    product['price']
    .astype(str)
    .str.replace('[^0-9.]', '', regex=True)  
    .str.strip()
)


product['price'] = pd.to_numeric(product['price'], errors='coerce')
```

```python
product['price'] =( 

pd.to_numeric(product['price'], errors='coerce')
.dropna()
)
```

```python
product = product[~product['product_name'].str.contains(r'[\?\؟]', regex=True, na=False)]

```

```python
product['product_name']=product['product_name'].astype(str).str.strip()
```

```python
product.drop_duplicates(inplace=True)
```

```python
product.sample(5)
```

```python
#لقيت مشكله بالبيانات بعضهم مخزنين ك اوبجكت ف المكتبه ماقدرت تتعامل معهم لي ساعتين ادور المشكله هههه

product.dtypes
```

```python
product['product_name']=product['product_name'].astype('string')
```

```python
product['product_id']=product['product_id'].astype('string').str.replace('"','').astype('int64')
```

```python
product.dropna(inplace=True)
```

```python
product.fillna(0 , inplace=True)
```

```python
product.count()
```

```python
product.dtypes
```

```python
product.reset_index(inplace=True)
```

```python
product.drop('index' , axis=1, inplace=True)
```

```python
product['product_name']=product['product_name'].str.strip('"')
```

```python
product = product[product["product_name"].str.fullmatch(r"[A-Za-z0-9\s\-]+")]
```
